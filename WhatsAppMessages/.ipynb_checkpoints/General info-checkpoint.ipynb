```python
import sys
import os
import pandas as pd
import numpy as np
import re


def read_file(filename):
    file = open(filename, mode='r', encoding='utf-8')
    data = file.read()
    file.close()   
    return data

def make_dataframe(data):
    messages = re.findall('(\d+/\d+/\d+, \d+:\d+:\d+ [A-Z]*]) (.*?:) (.*?\[)', data, flags=re.DOTALL)
    temp_df = pd.DataFrame(messages, columns=['date', 'name', 'message'])
    return temp_df


folder_path = './Chats'
df = pd.DataFrame(columns=['date', 'name', 'message', 'chat group name'])

for f in os.listdir(folder_path):
    f_path = folder_path + '/' + f
    data = read_file(f_path)
    temp_df = make_dataframe(data)
    temp_df['chat group name'] = f
    print(temp_df['name'][0])
    df = df.append(temp_df)
```

```python
df_msg_1 = pd.pivot_table(df, index=['message'], aggfunc=pd.Series.count)
df_msg_1
```

```python
from string_grouper import group_similar_strings

calls = pd.DataFrame(df_msg_1.index)
segments_pd = calls['message']
groupings = group_similar_strings(segments_pd, min_similarity=0.85)
groupings = pd.DataFrame(groupings)
groupings[1] = 1

df_msg_2 = pd.pivot_table(groupings, index=['message'], aggfunc=pd.Series.count)
df_msg_2
```
