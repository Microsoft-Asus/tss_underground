```python
import pandas as pd
import numpy as np

#pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_colwidth', 50)

df = pd.read_csv('posts.tsv', sep='\t', header=0)
print(df.shape)
df.head()
```

```python
import re

def na(msg):
    not_available = 'na' in msg
    if not_available:
         return True
    return False

def money_laundering(msg):
    blocking = ('blocking' in msg and 'payment' in msg)
    giftcard = ('gift card' in msg or 'giftcard' in msg or 'gifts cards' in msg) 
    zelle = 'zelle' in msg
    gateway_2d = ('2d' in msg and 'gateway' in msg)
    gateway_3d = ('3d' in msg and 'gateway' in msg)
    
    bank = 'bank account' in msg
    
    
    if blocking or giftcard or zelle or gateway_2d or gateway_3d or bank:
        return True
    return False


def website_developer(msg):
    call = ('call' in msg) or ('cll' in msg)
    available = 'available' in msg
    call_twice = (len(re.findall('call', msg)) > 1)
                   
    if call_twice or (call and available):
        return True
    return False

                   
def call_center_for_calls(msg):
    call = ('call' in msg or 'cll' in msg)
    need = 'need' in msg
    
    if call and need:
        return True
    return False


def call_center_for_agents(msg):
    looking = ('looking'in msg and 'agent' in msg)
    hiring = ('hiring' in msg and 'agent' in msg)
    
    if looking or hiring:
        return True
    return False    


def ppc(msg):
    adwords = ('adwords' in msg)
    ppc = ('ppc' in msg)
    
    if adwords or ppc:
        return True
    return False


def victim_data(msg):
    email_lead = 'email lead' in msg
    refund_lead = 'refund lead' in msg
    ivr_data = 'ivr data' in msg
    
    if email_lead or refund_lead or ivr_data:
        return True
    return False


def blasting(msg):
    email_blasting = ('email' in msg and 'blast' in msg)
    ivr_blasting = ('ivr' in msg and 'blast' in msg)
    
    if email_blasting or ivr_blasting:
        return True
    return False


def tfn(msg):
    toll_free_number = ('tfn available' in msg or 'toll free available' in msg 
                        or 'toll free number available' in msg)
    
    if toll_free_number:
        return True
    return False



def create_tag(msg):
    tag_text = []
    if na(msg):
        tag_text.append('empty post')
    
    if money_laundering(msg):
        tag_text.append('money laundering')
        
    if website_developer(msg):
        tag_text.append('website developer')
        
    if call_center_for_calls(msg):
        tag_text.append('call center for calls')
    
    if call_center_for_agents(msg):
        tag_text.append('call center for agents')
    
    if ppc(msg):
        tag_text.append('ppc')
            
    if victim_data(msg):
        tag_text.append('victim data')
    
    if blasting(msg):
        tag_text.append('blasting')
        
    if tfn(msg):
        tag_text.append('toll free number')
        
    if tag_text:
        return ','.join(tag_text)
    
    return 'other'
    
    
df['tag'] = df['message_text'].apply(create_tag)
```

```python
df[df['tag'] == 'other']
```

```python
tag_other = df[df['tag'] == 'other']
tag_other = pd.DataFrame(pd.pivot_table(tag_other, index=['message_text'], aggfunc=pd.Series.count).index)
tag_other.to_csv('other_tagged.csv', sep=',', header=True, index=False)
```
