```python
import pandas as pd
import numpy as np

#pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_colwidth', 50)

df = pd.read_csv('posts.tsv', sep='\t', header=0)
print(df.shape)
df.head()
```

```python
df['group_name'] = df['response_id'].map(lambda x: x.split('/')[0])
```

```python
pd.pivot_table(df, index=['actor_id'], aggfunc=pd.Series.count)
```

```python
df_group = pd.pivot_table(df, index=['group_name'], aggfunc=pd.Series.count)
df_group_actor_id = pd.pivot_table(df, index=['group_name','actor_id'], aggfunc=pd.Series.count)
df_group_actor_id
```

```python
for group in df_group.index:
    print(group)
    c = 0
    for (g, a_id) in df_group_actor_id.index:
        if g == group:
            c += 1
    print(c)
    print()
```

```python
df_actor_id_group = pd.pivot_table(df, index=['actor_id','group_name'], aggfunc=pd.Series.count)
df_actor_id = pd.pivot_table(df, index=['actor_id'], aggfunc=pd.Series.count)

actor_ids, num_groups = [], []
for actor_id in df_actor_id.index:
    actor_ids.append(actor_id)
    c = 0
    for (a_id, g) in df_actor_id_group.index:
        if a_id == actor_id:
            c += 1
    num_groups.append(c)
    

    
```

```python
df_actor_id
```

```python
df_actor_id_num_groups = pd.DataFrame(columns=['actor_id', 'num_groups'])
df_actor_id_num_groups['actor_id'] = actor_ids
df_actor_id_num_groups['num_groups'] = num_groups
pd.pivot_table(df_actor_id_num_groups, index=['num_groups'], aggfunc=pd.Series.count)
```
