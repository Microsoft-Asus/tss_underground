```python
import pandas as pd
import numpy as np

#pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_colwidth', 50)

df = pd.read_csv('../posts.tsv', sep='\t', header=0)
print(df.shape)
#df.head()
```

```python
def targeted_age(msg):
    if 'age filter' in msg or 'age target' in msg or 'old age' in msg:
         return True
    return False

df['targeted_age'] = df['message_text'].apply(targeted_age)
temp = df[df['targeted_age'] == True]

print(temp.shape)

temp = pd.DataFrame(pd.pivot_table(temp, index=['message_text'], aggfunc=pd.Series.count).index)
print(temp.shape)

temp.to_csv('age.txt', sep=',', header=True, index=False)
```

```python
def calls_starting(msg):
    #'calls running'
    if 'calls available' in msg or 'calls running' in msg or 'calls live' in msg:
         return True
    return False

df['calls_starting'] = df['message_text'].apply(calls_starting)
calls_starting = df[df['calls_starting'] == True]

print(calls_starting.shape)

temp = pd.DataFrame(pd.pivot_table(calls_starting, index=['message_text'], aggfunc=pd.Series.count).index)
print(temp.shape)

temp.to_csv('calls_starting.txt', sep=',', header=True, index=False)
```

```python
import time

def get_day(timestamp):
    return time.strftime("%a", time.localtime(timestamp))
    
def get_date(timestamp):
    return time.strftime("%Y-%m-%d", time.localtime(timestamp))
    
def get_time(timestamp):
    #return time.strftime("%H:%M:%S", time.localtime(timestamp))
    return time.strftime("%H", time.localtime(timestamp))

datetime = pd.DataFrame()
datetime['day'] = calls_starting['timestamp'].apply(get_day)
datetime['date']  = calls_starting['timestamp'].apply(get_date)
datetime['time']  = calls_starting['timestamp'].apply(get_time)
```

```python
pd.pivot_table(datetime, index=['day'], aggfunc=pd.Series.count)
```

```python
pd.pivot_table(datetime, index=['time'], aggfunc=pd.Series.count)
```
